```python
# ===============================
# 0️⃣ IMPORTS
# ===============================
import pandas as pd
import numpy as np
from prophet import Prophet
import optuna
from sklearn.metrics import mean_absolute_error
import matplotlib.pyplot as plt
import seaborn as sns

# ===============================
# 1️⃣ LOAD DATA
# ===============================
df = pd.read_csv('final_df_binary.csv', parse_dates=['date'])

# Make sure columns are numeric
df['total_task_time_minutes'] = df['total_task_time_minutes'].astype(float)
df['is_weekend'] = df['is_weekend'].astype(int)
df['is_holiday'] = df['is_holiday'].astype(int)

# Split train/test: train 2021-2023, test 2024
train = df[df['date'] < '2024-01-01'].copy()
test = df[df['date'].dt.year == 2024].copy()

# ===============================
# 2️⃣ DEFINE FUNCTION TO CREATE LAGS
# ===============================
def create_lags(df_sec):
    df_sec = df_sec.sort_values('date')
    df_sec['lag_1'] = df_sec['total_task_time_minutes'].shift(1).fillna(0)
    df_sec['lag_7'] = df_sec['total_task_time_minutes'].shift(7).fillna(0)
    df_sec['rolling_7'] = df_sec['total_task_time_minutes'].rolling(7, min_periods=1).mean()
    return df_sec

# ===============================
# 3️⃣ DEFINE OPTUNA OBJECTIVE FOR PROPHET
# ===============================
def objective(trial, df_sec):
    cps = trial.suggest_float('changepoint_prior_scale', 0.01, 0.9)
    sps = trial.suggest_float('seasonality_prior_scale', 1, 20)
    
    m = Prophet(
        daily_seasonality=True,
        yearly_seasonality=True,
        changepoint_prior_scale=cps,
        seasonality_prior_scale=sps
    )
    # Add regressors
    for reg in ['lag_1','lag_7','rolling_7','is_weekend','is_holiday']:
        m.add_regressor(reg)
    
    df_prophet = df_sec.rename(columns={'date':'ds','total_task_time_minutes':'y'})
    
    train_len = int(len(df_prophet)*0.8)
    df_train = df_prophet.iloc[:train_len]
    df_val = df_prophet.iloc[train_len:]
    
    m.fit(df_train)
    forecast = m.predict(df_val)
    
    y_true = df_val['y'].values
    y_pred = forecast['yhat'].values
    return mean_absolute_error(y_true, y_pred)

# ===============================
# 4️⃣ SECTION-WISE MODELING + HYPERPARAMETER TUNING
# ===============================
sections = df['section_id'].unique()
section_dfs = {}
section_stats_train = {}
section_stats_test = {}

for sec in sections:
    print(f"Processing {sec} ...")
    
    # 4.1 Filter section & create lags
    df_sec_train = create_lags(train[train['section_id']==sec].copy())
    df_sec_test = create_lags(test[test['section_id']==sec].copy())
    
    # 4.2 Optuna tuning
    study = optuna.create_study(direction='minimize')
    study.optimize(lambda trial: objective(trial, df_sec_train), n_trials=20)
    best_params = study.best_params
    print(f"Best params for {sec}: {best_params}")
    
    # 4.3 Fit final Prophet model
    m = Prophet(
        daily_seasonality=True,
        yearly_seasonality=True,
        changepoint_prior_scale=best_params['changepoint_prior_scale'],
        seasonality_prior_scale=best_params['seasonality_prior_scale']
    )
    for reg in ['lag_1','lag_7','rolling_7','is_weekend','is_holiday']:
        m.add_regressor(reg)
    
    df_train_prophet = df_sec_train.rename(columns={'date':'ds','total_task_time_minutes':'y'})
    m.fit(df_train_prophet)
    
    # 4.4 Predict on test
    df_test_prophet = df_sec_test.rename(columns={'date':'ds'})
    forecast = m.predict(df_test_prophet)
    df_sec_test['predicted_task_minutes'] = forecast['yhat'].values
    df_sec_test['error'] = df_sec_test['predicted_task_minutes'] - df_sec_test['total_task_time_minutes']
    section_dfs[sec] = df_sec_test[['date','section_id','is_weekend','is_holiday','total_task_time_minutes','predicted_task_minutes','error']].copy()
    
    # 4.5 Section-wise train stats
    train_stats = {
        'mean': df_sec_train['total_task_time_minutes'].mean(),
        'median': df_sec_train['total_task_time_minutes'].median(),
        'mode': df_sec_train['total_task_time_minutes'].mode()[0],
        'max': df_sec_train['total_task_time_minutes'].max(),
        'min': df_sec_train['total_task_time_minutes'].min(),
        'std': df_sec_train['total_task_time_minutes'].std(),
        'mae': mean_absolute_error(df_sec_train['total_task_time_minutes'], df_sec_train['total_task_time_minutes'].shift(1).fillna(0))
    }
    section_stats_train[sec] = train_stats
    
    # 4.6 Section-wise test stats
    test_stats = {
        'mean': df_sec_test['total_task_time_minutes'].mean(),
        'median': df_sec_test['total_task_time_minutes'].median(),
        'mode': df_sec_test['total_task_time_minutes'].mode()[0],
        'max': df_sec_test['total_task_time_minutes'].max(),
        'min': df_sec_test['total_task_time_minutes'].min(),
        'std': df_sec_test['total_task_time_minutes'].std(),
        'mae': mean_absolute_error(df_sec_test['total_task_time_minutes'], df_sec_test['predicted_task_minutes'])
    }
    section_stats_test[sec] = test_stats
    
# ===============================
# 5️⃣ CREATE SINGLE DF FOR ALL SECTIONS (OPTIONAL)
# ===============================
final_df = pd.concat(section_dfs.values()).sort_values(['section_id','date']).reset_index(drop=True)

# ===============================
# 6️⃣ VISUALIZATION
# ===============================
for sec in sections:
    df_plot = section_dfs[sec]
    plt.figure(figsize=(12,5))
    plt.plot(df_plot['date'], df_plot['total_task_time_minutes'], label='Actual', marker='o')
    plt.plot(df_plot['date'], df_plot['predicted_task_minutes'], label='Predicted', marker='x')
    plt.title(f'{sec} - Actual vs Predicted Task Minutes (2024)')
    plt.xlabel('Date')
    plt.ylabel('Task Minutes')
    plt.legend()
    plt.show()

# ===============================
# 7️⃣ SECTION-WISE STAT TABLES
# ===============================
train_stats_df = pd.DataFrame(section_stats_train).T
test_stats_df = pd.DataFrame(section_stats_test).T

print("Train Stats per Section:")
print(train_stats_df)
print("\nTest Stats per Section:")
print(test_stats_df)
```

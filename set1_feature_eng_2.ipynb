```python
# Calculate and display correlation between employees_on_duty and total_task_time_minutes for each section
section_corr = df.groupby('section_id')[['employees_on_duty', 'total_task_time_minutes']].corr().iloc[0::2, -1].reset_index()
section_corr = section_corr.rename(columns={'total_task_time_minutes': 'correlation_employees_vs_task_time'})
print(section_corr[['section_id', 'correlation_employees_vs_task_time']])
```

```python
# Visualize section-wise total_task_time_minutes over date
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(16,8))
sns.lineplot(data=df, x='date', y='total_task_time_minutes', hue='section_id', marker='o')
plt.title('Section-wise Total Task Time Minutes Over Date')
plt.xlabel('Date')
plt.ylabel('Total Task Time (Minutes)')
plt.legend(title='Section ID', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Plot separate diagrams for each section: date vs total_task_time_minutes
unique_sections = df['section_id'].unique()
for section in unique_sections:
    plt.figure(figsize=(12,6))
    section_df = df[df['section_id'] == section]
    plt.plot(section_df['date'], section_df['total_task_time_minutes'], marker='o')
    plt.title(f'Section {section}: Total Task Time Minutes Over Date')
    plt.xlabel('Date')
    plt.ylabel('Total Task Time (Minutes)')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

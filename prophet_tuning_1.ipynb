```python
# ===============================
# 1 IMPORTS
# ===============================
import pandas as pd
import numpy as np
from prophet import Prophet
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import itertools
import warnings
warnings.filterwarnings('ignore')

# ===============================
# 2 LOAD DATA
# ===============================
df = pd.read_csv('final_df_binary.csv')
df['date'] = pd.to_datetime(df['date'])
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['dayofweek'] = df['date'].dt.dayofweek
df['quarter'] = df['date'].dt.quarter
df['is_working_day'] = ((df['is_holiday']==0) & (df['is_weekend']==0)).astype(int)

train_df = df[df['year'] <= 2023].copy()
test_df = df[df['year'] == 2024].copy()

# ===============================
# 3 HYPERPARAMETER GRID
# ===============================
param_grid = {
    'changepoint_prior_scale': [0.01, 0.05, 0.1, 0.2, 0.03, 0.001],
    'seasonality_prior_scale': [1.0, 5.0, 8.0, 7.0, 30, 365, 100],
    'fourier_order_monthly': [3,5,7],
    'fourier_order_quarterly': [2,3,4]
}

# ===============================
# ===============================
# 4 FORECASTING LOOP (WITH TUNING + EXTRA REGRESSORS)
# ===============================
task_forecasts = {}

for section in train_df['section_id'].unique():
    section_train = train_df[(train_df['section_id']==section) & (train_df['is_working_day']==1)].copy()
    section_test = test_df[(test_df['section_id']==section) & (test_df['is_working_day']==1)].copy()
    
    if len(section_train) < 50 or len(section_test) == 0:
        print(f"Skipping section {section} due to insufficient data")
        continue
    
    # Prepare Prophet df
    prophet_df = section_train[['date','total_task_time_minutes','is_holiday','is_weekend']].rename(
        columns={'date':'ds','total_task_time_minutes':'y'})
    
    best_mae = float('inf')
    best_model = None
    
    # Grid search
    for cps, sps, fourier_m, fourier_q in itertools.product(
        param_grid['changepoint_prior_scale'],
        param_grid['seasonality_prior_scale'],
        param_grid['fourier_order_monthly'],
        param_grid['fourier_order_quarterly']
    ):
        model = Prophet(
            yearly_seasonality=True,
            weekly_seasonality=False,
            daily_seasonality=False,
            changepoint_prior_scale=cps,
            seasonality_prior_scale=sps
        )
        model.add_seasonality(name='monthly', period=30.5, fourier_order=fourier_m)
        model.add_seasonality(name='quarterly', period=91.25, fourier_order=fourier_q)
        
        # Add extra regressors
        model.add_regressor('is_holiday')
        model.add_regressor('is_weekend')
        
        model.fit(prophet_df)
        
        future_df = prophet_df[['ds','is_holiday','is_weekend']]
        forecast = model.predict(future_df)
        forecast['yhat'] = forecast['yhat'].clip(lower=0)
        
        mae = mean_absolute_error(prophet_df['y'], forecast['yhat'])
        if mae < best_mae:
            best_mae = mae
            best_model = model
    
    print(f"{section}: Best MAE={best_mae:.2f}, CPS={best_model.changepoint_prior_scale}, SPS={best_model.seasonality_prior_scale}")
    
    # Forecast 2024 with regressors
    future_2024 = section_test[['date','is_holiday','is_weekend']].rename(columns={'date':'ds'})
    forecast_2024 = best_model.predict(future_2024)
    forecast_2024['yhat'] = forecast_2024['yhat'].clip(lower=0)
    
    # Save forecast with section_id
    task_forecasts[section] = pd.DataFrame({
        'date': section_test['date'],
        'section_id': section_test['section_id'],
        'actual_task_minutes': section_test['total_task_time_minutes'],
        'predicted_task_minutes': forecast_2024['yhat'].values
    })

# ===============================
# 5 COMBINE FORECASTS
# ===============================
combined_forecast = pd.concat(task_forecasts.values(), axis=0)


# ===============================
# 6 EVALUATION METRICS
# ===============================
print("\n=== OVERALL PERFORMANCE ===")
mae_overall = mean_absolute_error(combined_forecast['actual_task_minutes'], combined_forecast['predicted_task_minutes'])
rmse_overall = np.sqrt(mean_squared_error(combined_forecast['actual_task_minutes'], combined_forecast['predicted_task_minutes']))
r2_overall = r2_score(combined_forecast['actual_task_minutes'], combined_forecast['predicted_task_minutes'])
print(f"MAE: {mae_overall:.2f}, RMSE: {rmse_overall:.2f}, R: {r2_overall:.3f}")

# Section-wise
section_results = []
for section in combined_forecast['section_id'].unique():
    sec_data = combined_forecast[combined_forecast['section_id']==section]
    mae = mean_absolute_error(sec_data['actual_task_minutes'], sec_data['predicted_task_minutes'])
    rmse = np.sqrt(mean_squared_error(sec_data['actual_task_minutes'], sec_data['predicted_task_minutes']))
    r2 = r2_score(sec_data['actual_task_minutes'], sec_data['predicted_task_minutes'])
    section_results.append({
        'Section': section, 'MAE': mae, 'RMSE': rmse, 'R': r2
    })
section_results_df = pd.DataFrame(section_results)
print("\n=== SECTION-WISE PERFORMANCE ===")
print(section_results_df)

# ===============================
# 7 VISUALIZATIONS
# ===============================
plt.figure(figsize=(10,6))
sns.scatterplot(x='actual_task_minutes', y='predicted_task_minutes', hue='section_id', data=combined_forecast, alpha=0.7)
plt.plot([0, combined_forecast['actual_task_minutes'].max()],
         [0, combined_forecast['actual_task_minutes'].max()],
         'r--', linewidth=2)
plt.xlabel('Actual Task Minutes')
plt.ylabel('Predicted Task Minutes')
plt.title('Actual vs Predicted Task Minutes (All Sections)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

# Section-wise scatter plots
sections = combined_forecast['section_id'].unique()
for section in sections:
    sec_data = combined_forecast[combined_forecast['section_id']==section]
    plt.figure(figsize=(8,5))
    plt.scatter(sec_data['actual_task_minutes'], sec_data['predicted_task_minutes'], alpha=0.6)
    plt.plot([0, sec_data['actual_task_minutes'].max()],
             [0, sec_data['actual_task_minutes'].max()], 'r--', linewidth=2)
    plt.xlabel('Actual Task Minutes')
    plt.ylabel('Predicted Task Minutes')
    plt.title(f'{section} - Actual vs Predicted')
    plt.grid(True, alpha=0.3)
    plt.show()

# Monthly trends comparison per section
combined_forecast['year_month'] = combined_forecast['date'].dt.to_period('M')
monthly_comparison = combined_forecast.groupby(['year_month','section_id']).agg({
    'actual_task_minutes':'mean',
    'predicted_task_minutes':'mean'
}).reset_index()

for section in sections:
    sec_monthly = monthly_comparison[monthly_comparison['section_id']==section]
    plt.figure(figsize=(10,5))
    plt.plot(sec_monthly['year_month'].astype(str), sec_monthly['actual_task_minutes'], 'o-', label='Actual', linewidth=2)
    plt.plot(sec_monthly['year_month'].astype(str), sec_monthly['predicted_task_minutes'], 's-', label='Predicted', linewidth=2)
    plt.title(f'{section} - Monthly Average Task Minutes 2024')
    plt.xlabel('Month')
    plt.ylabel('Avg Task Minutes')
    plt.xticks(rotation=45)
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.show()
```

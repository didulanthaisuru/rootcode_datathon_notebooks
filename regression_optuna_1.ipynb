```python
# Section-wise Regression Model Parameter Tuning with Optuna
# This script tunes HuberRegressor parameters (`epsilon`, `alpha`) for each section to minimize MAE.

import pandas as pd
import numpy as np
from sklearn.linear_model import HuberRegressor
from sklearn.metrics import mean_absolute_error
import optuna
import matplotlib.pyplot as plt
import seaborn as sns

# Load and preprocess data
df = pd.read_csv('final_df_binary.csv')
df['date'] = pd.to_datetime(df['date'])
df['year'] = df['date'].dt.year
df['is_working_day'] = ((df['is_holiday'] == 0) & (df['is_weekend'] == 0)).astype(int)
train_data = df[(df['year'] <= 2023) & (df['is_working_day'] == 1)].copy()

best_params_dict = {}
results = []
sections = train_data['section_id'].unique()

for section in sections:
    section_data = train_data[train_data['section_id'] == section]
    if len(section_data) < 30:
        continue
    X = section_data[['total_task_time_minutes']].values
    y = section_data['employees_on_duty'].values

    def objective(trial):
        epsilon = trial.suggest_float('epsilon', 1.0, 4.0)
        alpha = trial.suggest_float('alpha', 0.000001, 1.0, log=True)
        model = HuberRegressor(epsilon=epsilon, alpha=alpha)
        model.fit(X, y)
        preds = model.predict(X)
        mae = mean_absolute_error(y, preds)
        return mae

    study = optuna.create_study(direction='minimize')
    study.optimize(objective, n_trials=90, show_progress_bar=False)
    best_params = study.best_params
    best_params_dict[section] = best_params
    results.append({
        'section_id': section,
        'best_epsilon': best_params['epsilon'],
        'best_alpha': best_params['alpha'],
        'best_mae': study.best_value
    })
    print(f"Section {section}: Best epsilon={best_params['epsilon']:.3f}, alpha={best_params['alpha']:.5f}, MAE={study.best_value:.3f}")

results_df = pd.DataFrame(results)
results_df.to_csv('regression_tuning_results.csv', index=False)

plt.figure(figsize=(10,6))
sns.barplot(x='section_id', y='best_mae', data=results_df)
plt.title('Best MAE by Section (Regression Tuning)')
plt.xlabel('Section')
plt.ylabel('Best MAE')
plt.show()

print('Tuning results saved to regression_tuning_results.csv')
```
